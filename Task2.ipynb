```python
# TASK 2 
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('titanic.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Cabin'].fillna('Unknown', inplace=True)
data['Embarked'].fillna('Unknown', inplace=True)
```

```python
data.isnull().sum()
```

```python

data.hist(bins=10, figsize=(10, 6))

# Adding labels and title
plt.xlabel('Variable')
plt.ylabel('Frequency')
plt.title('Histogram for Data')

# Adjusting layout for better appearance
plt.tight_layout()

# Display the plot
plt.show()
```

```python
pd.plotting.scatter_matrix(data, figsize=(10, 6))
plt.show()
```

```python
data.boxplot(column='Fare', by='Pclass')
plt.show()
```

```python
data.fillna('', inplace=True) 
data.head()
```

```python
numeric_data = data.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()

# Display the correlation matrix
print(correlation_matrix)
```

```python

data['Fare'].hist()

# Adding labels and title for better readability
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Histogram of Fare')

# Display the plot
plt.show()
```

```python
data.boxplot(column='Fare')
```

```python
plt.scatter(data['Age'], data['Fare'])
```

```python
sns.pairplot(data)
```

```python

sns.countplot(x='Sex', data=data)

# Adding labels and title for better readability
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Countplot of Sex')

# Display the plot
plt.show()
```
